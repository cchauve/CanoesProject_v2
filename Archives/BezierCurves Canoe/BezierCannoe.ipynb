```python
def Lerp(A,B,t):
    return [A[0]*(1-t)+B[0]*t, A[1]*(1-t)+B[1]*t]
```

```python
def BezierFirstStep(A = [1,4], B=[3,1], C = [4,6], D = [6,3], t=0.5):
    AB = Lerp(A, B, t)
    BC = Lerp(B, C, t)
    CD = Lerp(C, D, t)
    
    fig, ax = plt.subplots()
    
    ax.plot([A[0],B[0]],[A[1],B[1]], 'k')
    ax.plot([B[0],C[0]],[B[1],C[1]], 'k')
    ax.plot([C[0],D[0]],[C[1],D[1]], 'k')
    
    ax.plot(A[0], A[1], 'ko')
    ax.plot(B[0], B[1], 'ko')
    ax.plot(C[0], C[1], 'ko')
    ax.plot(D[0], D[1], 'ko')
    
    ax.plot(AB[0], AB[1], 'bo')
    ax.plot(BC[0], BC[1], 'bo')
    ax.plot(CD[0], CD[1], 'bo')

    ax.annotate("A", (A[0],A[1]), textcoords = "offset points", xytext=(0,10), ha = 'center')
    ax.annotate("B", (B[0],B[1]), textcoords = "offset points", xytext=(0,10), ha = 'center')
    ax.annotate("C", (C[0],C[1]), textcoords = "offset points", xytext=(0,-20), ha = 'center')
    ax.annotate("D", (D[0],D[1]), textcoords = "offset points", xytext=(0,10), ha = 'center')
    
    plt.show()
```

```python
def BezierThirdStep(A = [1,4], B=[3,1], C = [4,6], D = [6,3], t=0.5):
    
    AB = Lerp(A , B , t)
    BC = Lerp(B , C , t)
    CD = Lerp(C , D , t)
    U  = Lerp(AB, BC, t)
    V  = Lerp(BC, CD, t)
    UV = Lerp(U , V , t)
    
    fig, ax = plt.subplots()
    
    ax.plot([A[0],B[0]],[A[1],B[1]], 'k')
    ax.plot([B[0],C[0]],[B[1],C[1]], 'k')
    ax.plot([C[0],D[0]],[C[1],D[1]], 'k')
    ax.plot([AB[0],BC[0]],[AB[1],BC[1]], 'b')
    ax.plot([BC[0],CD[0]],[BC[1],CD[1]], 'b')
    ax.plot([U[0],V[0]],[U[1],V[1]], 'm')
    
    
    
    ax.plot(A[0], A[1], 'ko')
    ax.plot(B[0], B[1], 'ko')
    ax.plot(C[0], C[1], 'ko')
    ax.plot(D[0], D[1], 'ko')
    
    ax.plot(AB[0], AB[1], 'bo')
    ax.plot(BC[0], BC[1], 'bo')
    ax.plot(CD[0], CD[1], 'bo')
    
    ax.plot(U[0], U[1], 'ro')
    ax.plot(V[0], V[1], 'ro')
    
    ax.plot(UV[0], UV[1], 'mo')

    ax.annotate("A", (A[0],A[1]), textcoords = "offset points", xytext=(0,10), ha = 'center')
    ax.annotate("B", (B[0],B[1]), textcoords = "offset points", xytext=(0,10), ha = 'center')
    ax.annotate("C", (C[0],C[1]), textcoords = "offset points", xytext=(0,-20), ha = 'center')
    
    plt.show()
```

```python
def BezierFourthStep(A = [1,4], B=[3,1], C = [4,6], D = [6,3], t=0.5 , N=64):
    
    T = np.arange(N)/(N-1)
    X = [0] * N
    Y = [0] * N
    
    for i in range(0, N):
        AB = Lerp(A , B , T[i])
        BC = Lerp(B , C , T[i])
        CD = Lerp(C , D , T[i])
        U  = Lerp(AB, BC, T[i])
        V  = Lerp(BC, CD, T[i])
        UV = Lerp(U , V , T[i])
 
        X[i] = UV[0]
        Y[i] = UV[1]
    
    
    AB = Lerp(A , B , t)
    BC = Lerp(B , C , t)
    CD = Lerp(C , D , t)
    U  = Lerp(AB, BC, t)
    V  = Lerp(BC, CD, t)
    UV = Lerp(U , V , t)
    
    fig, ax = plt.subplots()
    
    
    ax.plot([A[0],B[0]]  ,[A[1],B[1]]  , 'k', linewidth = 1.0)
    ax.plot([B[0],C[0]]  ,[B[1],C[1]]  , 'k', linewidth = 1.0)
    ax.plot([C[0],D[0]]  ,[C[1],D[1]]  , 'k', linewidth = 1.0)
    ax.plot([AB[0],BC[0]],[AB[1],BC[1]], 'b', linewidth = 1.0)
    ax.plot([BC[0],CD[0]],[BC[1],CD[1]], 'b', linewidth = 1.0)
    ax.plot([U[0],V[0]]  ,[U[1],V[1]]  , 'm', linewidth = 1.0)
    
    ax.plot(X,Y, 'm', linewidth = 2.0)
    
    
    ax.plot(A[0], A[1], 'ko')
    ax.plot(B[0], B[1], 'ko')
    ax.plot(C[0], C[1], 'ko')
    ax.plot(D[0], D[1], 'ko')
    
    ax.plot(AB[0], AB[1], 'bo')
    ax.plot(BC[0], BC[1], 'bo')
    ax.plot(CD[0], CD[1], 'bo')
    
    ax.plot(U[0], U[1], 'ro')
    ax.plot(V[0], V[1], 'ro')
    
    ax.plot(UV[0], UV[1], 'mo')

    ax.annotate("A", (A[0],A[1]), textcoords = "offset points", xytext=(0,10), ha = 'center')
    ax.annotate("B", (B[0],B[1]), textcoords = "offset points", xytext=(0,10), ha = 'center')
    ax.annotate("C", (C[0],C[1]), textcoords = "offset points", xytext=(0,-20), ha = 'center')
    
    plt.show()
```

```python
class BezierCurvePart:
    
    A = [0]*2
    B = [0]*2
    C = [0]*2
    D = [0]*2
    
    def __init__(self, A, B, C, D):
        self.A = A
        self.B = B
        self.C = C
        self.D = D
    
    def __Lerp(A,B,t):
        return [A[0]*(1-t)+B[0]*t, A[1]*(1-t)+B[1]*t]
    
    def Get(t):
        X = A[0]*Weight(t, 1) + B[0] * Weight(t, 2) + C[0] * weight(t,3) + D[0]*weight(t,4)
        Y = A[1]*Weight(t, 1) + B[1] * Weight(t, 2) + C[1] * weight(t,3) + D[1]*weight(t,4)
        return [X,Y]
    def GetPoint(n):
        if n == 1: return A
        if n == 2: return B
        if n == 3: return C
        if n == 4: return D
        return [0,0]
    
    def __Weight(t, n : int): 
        #switch case only in python 3.10+ (or in pythons situation Match case)
        if n == 1: return -1*pow(t,3) + 3*pow(t,2) - 3*t+1
        if n == 2: return  3*pow(t,3) - 6*pow(t,2) + 3*t
        if n == 3: return -3*pow(t,3) + 3*pow(t,2)
        if n == 4: return    pow(t,3)
        return 0
```
