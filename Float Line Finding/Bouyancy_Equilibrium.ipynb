```python
from scripts import plot_funcs as canoe
from scripts import equilibrium as eq
from matplotlib import pyplot as plt
import numpy as np
import plotly.graph_objs as go
```

```python
N = 64 #resolution

#TEST: Height = 24" = 0.61m, length = 16' = 4.8768m, Beam = 12" = 0.31m
W1 = [0.31]
W2 = [0.61]
R3 = [4.8768/2.0]
a = [0.15]

eta, theta = np.meshgrid(np.linspace(0.01, np.pi - 0.01, N), np.linspace(np.pi / 2, 3 * np.pi / 2, N))

X, Y, Z = canoe.type_three(W1[0], W2[0], R3[0], a, eta, theta)
fig = canoe.plot_canoe(R1=W1, R2=W2, R3=R3, a=a, eta=eta, theta=theta, type=3)
fig.update_layout(margin=dict(l=20, r=20, t=20, b=20), scene_aspectmode="cube")
fig.show()
```

```python
heightNormalArea = eq.GenerateVectorList(X,Y,Z)
area = eq.GetArea(heightNormalArea)
weight = np.arange(-200, 11800, 200)
N = len(weight)
zLevel = [0]*N
for i in range(N):
    zLevel[i] = -eq.BinarySearch(heightNormalArea, weight[i])

```

```python
fig = plt.figure()
fig.set_figheight(10)
fig.set_figwidth(30)
ax = fig.add_subplot(121)
ax.plot(weight, zLevel, 'r', linewidth = 2.0)
ax.set_xlabel('weight (Newtons)')
ax.set_ylabel('z level (meters)')
plt.show()
```

This 16 foot canoe has a absolute maximum holding capactiy of 1172 Kg, which translates roughly 11500 Newtons. Of course that would put the water line at the rim of the canoe which would not be ideal, a water line at z = -0.3 would be more ideal (halfway down the boat). This gives us a capacity of 407 Kg, which translates roughly to 4000 Newtons. This roughly matches with the maximum weight capacity of normal 16 foot canoes, roughly 426 Kg.
